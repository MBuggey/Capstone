Ideas to implement based on the information available within the three dataframes are as follows:

- Review age, we have a date that a review was posted.  We also have the date that a game was released available but due to the fact that early access games are quite popular and can be reviewed prior to official release that metric would have to many issues.  However, we can compare a review age to the last date available in the recommendations dataframe to determine review age.  
- is_recommended is the easiest factor to determine a weighting.  Making it the strongest weight will also ensure that ratings don't go too far off from the original data skew.
- Hours can be used as well, we can calculate the median playtime for each game and determine if a user played enough of a game to weight their opinion higher or lower.  This is mostly useful when taking into consideration bot accounts that purchase games, rate them, then refund them before the 2 hour refund limit.  
- We can also calculate a users review ratio (amount of products owned vs reviews posted) to determine if a user is a good representation of general concensus.  Potentially weighting people who review more (or less potentially) than normal and instilling a cut off range so individuals with low products and high reviews aren't weighted extremely highly.
- We can also use overall game rating and positivity ratio to determine whether a review aligns or does not align with overall user concensus.  
- Finally we can look at the total user_reviews in games.csv and determine the ratio of how many reviews of the total made it into our recommendations dataframe.  From there we can weight items in a manner to avoid over/underrepresentation. 

```python
# taking into consideration what I learned in sprint 1 that mostly positive and negative aren't as harsh I'm gonna remap the 
# scores so that overwhelming takes precedent and positive and negative encompass everything else.  These will be used as 
# weights to determine whether a review aligns with or against overall userbase sentiment on a game.

rating_mapping = {
        'Overwhelmingly Positive': 1.3,
        'Very Positive': 1.2,
        'Positive': 1.2,
        'Mostly Positive': 1.1,
        'Mixed': 1,
        'Mostly Negative': 0.8,
        'Negative': 0.6,
        'Very Negative': 0.6,
        'Overwhelmingly Negative': 0.2
}

def map_rating(value):
    for key, score in rating_mapping.items():
        if key.lower() in value.lower():
            return score
    return None  


games['rating'] = games['rating'].apply(map_rating)
```

After exploring the majority of what I can initially in the games and recommendations dataframe now I'm going to move on to users.  The goal is to build a review_ratio to determine how the average user rates.

```python
# making a new column in users which calculates the users review ratio and replaces any 0's with nan values
users['review_ratio'] = users['reviews'] / users['products'].replace(0, np.nan)
```

There was a lot of exploratory EDA which was removed but for reader reference I discovered that there were simultaneously a lot of users who had NaN review ratios alongside ones with review_ratios over 1. 

The end result was that there were a lot of bot users who were purchasing games, reviewing them, then refunding them before the 2hour refund limit was up. 

This lead to them having little to no products yet many reviews. There were also legitimate users however who signed up for family accounts which allows you to review a game without owning it.  However, there was no real automated way of determining bots vs family account holders and all users with abnormal review ratios only added up to less than 0.5% of total data so they were simply removed.

```python
# Removing nan users from recommendations and users

nan_users = users.loc[users['review_ratio'].isna()]

bot_filterer = recommendations[recommendations['user_id'].isin(nan_users['user_id'])]  

is_bot_user = recommendations['user_id'].isin(bot_filterer['user_id'])

recommendations = recommendations[~is_bot_user]

users.dropna(subset=['review_ratio'], inplace=True)
```

```python
# Removing users with review ratios greater than 1 from recommendations and users.

positive_rat_users = users[users['review_ratio'] > 1]

delete = recommendations['user_id'].isin(positive_rat_users['user_id'])

recommendations = recommendations[~delete]

users = users[users['review_ratio'] <= 1]
```

Back to the original point of determining the review ratio.  It is simultaneously a weight and a stopgap to determine whether a user meets or exceeds the average users tendency.  Thus the mean and median review ratios should be taken into account.  Users below, meeting, or slightly exceeding these values could be weighted slightly higher or normally depending on their ratio.  And it can also be used as a stopgap to prevent users who either own very few products and rate that small handful or users who rate a lot from dominating the algorithm.

```python
# Setting a review age column.
# last review date as of scrape was 2022-12-31, more recent reviews should be 
# weighted heavier than older ones.  However, that should also take into consideration
# each individual games rating and positive ratio 

last_review_date = pd.Timestamp('2022-12-31')

recommendations['review_age'] = (last_review_date - recommendations['date']).dt.days
```

```python
# Merging extra columns from users and games into recommendations to assist with the weighting function
# Additionally dropping any columns that I determined would not be helpful for the final calculation.

recommendations = recommendations.merge(users, on='user_id')
recommendations = recommendations.merge(games, on='app_id')

recommendations.drop(columns=['win','mac','linux','steam_deck','discount','price_final','price_original','date_release'],inplace=True)
```

```python
# Grouping all occurences of a unique app_id in recommendations and dividing it by the games total user_reviews and adding 
# it to a new column called game_review_ratio.  This is to determine if a game is over/underrepresented.

recommendations['game_review_ratio'] = recommendations.groupby('app_id')['is_recommended'].transform('count') / recommendations['user_reviews']
```

```python
# Large function to calculate a new composite rating for each review in recommendations

def calculate_ratings(df, 
                      
                      mean_review_ratio=0.06, 
                      
                      median_review_ratio=0.03, 
                      
                      max_user_reviews_cap=10000, # games with reviews over 10,000 won't be weighted more than others 
                      
                      min_games_threshold=60, # Threshhold inbetween mean and median to not weight users with high review ratios and low products highly
                      
                      boost_positive_ratio_threshold=80, # Games with positive ratios over 80 get weighted higher 
                      
                      boost_user_reviews_threshold=500, # Sprint 1 metric showed games with at least 500 reviews qualify for overwhelming scores
                      
                      cap_positive_ratio_threshold=70, # Used for capping games with lots of reviewers and not fantastic positive ratios
                      
                      low_game_review_ratio_threshold=0.1, # boosts or lessens weights based on game representation in df
                      
                      game_review_ratio_boost_factor=1.1 # same as above
                      
                     
                     ):
    
    weight_hours = 2 # weighted higher as hours are usually a strong indicator
    
    weight_helpful = 1 # helpful and funny weighted less because of how infrequent they are
    
    weight_funny = 1
    
    weight_is_recommended = 3 # the original positivity metric is the highest weighted 
    
    weight_review_ratio = 2 # Exemplary users are rated higher 
    
    weight_positive_ratio = 1 # I want to weight overall positive games a little higher but not by much
    
    weight_user_reviews = 1 # Same as above but for review numbers, good indicator but don't want it to dominate
    
    weight_game_rating = 2 # This can be a bit stronger as it is indicative of the games overall score 

    playtime_threshold = 15  # Users within 15 hours of average are not penalized

   
    # Normalize helpful and funny as they're a little odd
    max_log_helpful = df['log_helpful'].max()
    max_log_funny = df['log_funny'].max()

    
    # First boost conditions goal is to boost weights of games that meet or exceed an 80% positivty ratio and have at least 
    # 500 reviews and does not exceed 10,000.  10,000 could potentially be reduced later.  
    boost_condition = (df['positive_ratio'] >= boost_positive_ratio_threshold) & (df['user_reviews'] >= boost_user_reviews_threshold) & (df['user_reviews'] < 10000)
    
    # Caps games with lots of reviews and less then 70% positivity ratio from being weighted higher due to their presence
    cap_condition = (df['user_reviews'] >= max_user_reviews_cap) & (df['positive_ratio'] <= cap_positive_ratio_threshold)

    
    # Checks player game time compared to the average and see if it meets the threshhold  
    hour_weight = (df['hours'] - df['average_hours']).abs() <= playtime_threshold
    
    # Checks if a user has enough games to meet the minimum game threshhold and be considered an avg user 
    review_ratio_weight = df['products'] >= min_games_threshold
    
    # caps upper limit of user reviews to not disproportionately scale highly reviewed titles  
    user_review_weight = df['user_reviews'].clip(upper=max_user_reviews_cap) / max_user_reviews_cap

    # calculates the total weight value
    total_weights = weight_hours + weight_helpful + weight_funny + weight_is_recommended + weight_review_ratio + weight_positive_ratio + weight_user_reviews + weight_game_rating

    # calculates the composite rating 
    df['composite_rating'] = (
        (hour_weight * weight_hours) +
        (df['log_helpful'] / max_log_helpful * weight_helpful) +
        (df['log_funny'] / max_log_funny * weight_funny) +
        (df['is_recommended'] * weight_is_recommended) +
        (review_ratio_weight * df['review_ratio'] / max(mean_review_ratio, median_review_ratio) * weight_review_ratio) +
        (df['positive_ratio'] / 100 * weight_positive_ratio * np.where(boost_condition, 1.2, 1)) +
        (user_review_weight * weight_user_reviews * np.where(cap_condition, 0.8, 1)) +
        (df['rating'] * weight_game_rating)
    ) / total_weights * 10

    # Adjust rating based on game_review_ratio
    df['composite_rating'] *= np.where(df['game_review_ratio'] < low_game_review_ratio_threshold, game_review_ratio_boost_factor, 1)

    # Ensure rating is between 0 and 10 and not a float
    df['composite_rating'] = df['composite_rating'].clip(lower=0, upper=10)
    df['composite_rating'] = np.around(df['composite_rating'])
    df['composite_rating'] = df['composite_rating'].astype(int)

    return df


calculate_ratings(recommendations)  
```
