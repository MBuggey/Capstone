```python
import pandas as pd
from scipy.sparse import coo_matrix, vstack
import numpy as np
from tqdm import tqdm

# Determine the total number of unique items and users
total_items = pd.read_csv('../Data/recommendations.csv')['app_id'].nunique()
total_users = pd.read_csv('../Data/recommendations.csv')['user_id'].nunique()

# Reading the DataFrame in chunks
chunk_size = 10000
total_rows = pd.read_csv('../Data/recommendations.csv', usecols=[0]).shape[0]  # Get the total number of rows
num_chunks = np.ceil(total_rows / chunk_size).astype(int)  # Calculate the total number of chunks

chunks = pd.read_csv('../Data/recommendations.csv', chunksize=chunk_size)

final_sparse_matrix = None

for chunk in tqdm(chunks, total=num_chunks, desc="Processing Chunks"):
    # Convert 'is_recommended' to integers and factorize user_id and app_id
    user_ids = pd.factorize(chunk['user_id'])[0]
    app_ids = pd.factorize(chunk['app_id'])[0]
    interaction_values = chunk['is_recommended'].astype(np.int8)

    # Create a sparse matrix for the chunk with consistent dimensions
    chunk_sparse_matrix = coo_matrix((interaction_values, (user_ids, app_ids)), 
                                     shape=(total_users, total_items))

    # Combine with the final matrix
    if final_sparse_matrix is None:
        final_sparse_matrix = chunk_sparse_matrix
    else:
        final_sparse_matrix = vstack([final_sparse_matrix, chunk_sparse_matrix])

# 'final_sparse_matrix' is now your complete interaction matrix
```

```python
from scipy.sparse import vstack

# Define the chunk size - number of rows to include in each chunk
chunk_size = 10000  # Adjust this based on your system's capacity

# Calculate the number of chunks
num_chunks = interactions.shape[0] // chunk_size
if interactions.shape[0] % chunk_size != 0:
    num_chunks += 1

```

```python
# Convert the interactions matrix to CSR format
interactions_csr = interactions.tocsr()
```

```python
from lightfm import LightFM
from tqdm import tqdm

# Initialize the LightFM model
model = LightFM(no_components=30, learning_rate=0.05, loss='warp')

# Train the model incrementally on each chunk
for chunk_index in tqdm(range(num_chunks), desc="Training Model"):
    start_index = chunk_index * chunk_size
    end_index = min((chunk_index + 1) * chunk_size, interactions_csr.shape[0])

    # Slice the interaction matrix to get the current chunk
    interactions_chunk = interactions_csr[start_index:end_index]

    # Fit the model on this chunk
    model.fit_partial(interactions_chunk, 
                      user_features=user_features_matrix, 
                      item_features=item_features_matrix, 
                      epochs=1, 
                      num_threads=1)  # num_threads is set to 1 due to lack of OpenMP support
```
